```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import matplotlib.dates as mdates
import datetime as dt

# The path to our CSV file
file = "Resources/Metro_Nashville___Davidson_County_Traffic_Crashes__2019_.csv"

# Read our CSV data into pandas
df = pd.read_csv(file, parse_dates = ['Date and Time'])
df.head()
```

```python
#Get row count for each column
df.count()
```

```python
#Drop nan or blank only in date/time column

df = df.dropna(subset=["Date and Time", "Mapped Location"])

df.head()
```

```python
df['month_year'] = df['Date and Time'].dt.to_period('M')
df['Year'] = pd.DatetimeIndex(df['Date and Time']).year.astype(str)
df['Year'] = df['Year'].astype(str).replace('\.0', '', regex=True)
df['Month'] = pd.DatetimeIndex(df['Date and Time']).month.astype(str)
df['Month'] = df['Month'].astype(str).replace('\.0', '', regex=True)
```

```python
df.count()
```

```python
#df.astype({"Year" : "int64"})
#df[df.Year == {2010,2011,2012,2013,2014,2015,2016,2017,2018,2019}]
#df.head()
df = df[(df['Date and Time'].dt.year >= 2010)]
df = df[(df['Latitude']) != 0.0000]
df = df[(df['Longitude']) != 0.0000]
```

```python
df.to_csv(r'Resources/Metro_Crash_Clean.csv')
df.head()
```
