```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import matplotlib.dates as mdates
import datetime as dt
import os

#environment variable set int command/git bash with:  export gkey=keyfromgoogle
#pull in environment variable, neet to import os for this to work
gkey = os.getenv("gkey")

# The path to our CSV file
file = "Resources/Metro_Crash_Clean.csv"

# Read our CSV data into pandas
df = pd.read_csv(file, parse_dates = ['Date and Time'],low_memory=False)
df.count()
```

```python
##Must do for this in git bash / command line to run:
    #pip install gmaps
    #jupyter nbextension enable --py gmaps

#Initial setup help https://medium.com/future-vision/google-maps-in-python-part-2-393f96196eaf
#GMAPS documentation https://jupyter-gmaps.readthedocs.io/en/latest/

import gmaps

gmaps.configure(api_key=gkey)

#geo for center of nashville
nashville_center = (36.1960, -86.7994)

#defines how the image looks, can set height/width with number of pixels ex. 400px
figure_layout = {
    'width': '600px',
    'height': '600px',
    'border': '1px solid black',
    'padding': '1px'}

#defines locations ad the combination of the latitude and longitude fields
locations = df[["Latitude", "Longitude"]]

#sets the baseline figure
fig = gmaps.figure(map_type="ROADMAP", layout=figure_layout, center=nashville_center, zoom_level=10)
#map types are ROADMAP, SATELLITE, HYBRID, and TERRAIN

#builds heatmap off of the dataframe
heatmap_layer = gmaps.heatmap_layer(locations)
heatmap_layerb = gmaps.heatmap_layer(locations)

#adds heatmap to the figure
fig.add_layer(heatmap_layer)
fig.add_layer(heatmap_layerb)

#displays the figure
fig

```

```python
#heatmap settings can be modified after displaying in the cell above.  
heatmap_layer.max_intensity = 400
heatmap_layer.point_radius = 10
heatmap_layerb.max_intensity = 400
heatmap_layerb.point_radius = 10
heatmap_layerb.gradient = [(0,0,0,0),(0,255,0,0),(255,246,0,1),(255,0,0,1)]
#heatmap_layer.gradient = ["green","blue","red"]

#color gradiants tend to color whole map but look good.
heatmap_layer.opacity = .3
heatmap_layerb.opacity = 1
```
